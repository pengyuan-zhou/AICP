```python
import pandas as pd
df = pd.read_json(r'objects.json')
```

```python
df.head()
```

```python
import time
start_time = time.clock()
result = 0
rows=[]
objects_high=[]
objects_low=[]
objects=[]
labels=[]
label=0
for row in df.itertuples(index=True):
    for box in row:
        if isinstance(box, int):
            continue
        if box is None:
            continue
        tmp=[]
        tmp.append(box['distance'])
        tmp.append(box['velocity'])
        if box['distance']< 23 or box['velocity']> 30 or box['class']=='person':
            tmp.append(1) #replace person with 1, other classes with 0, for simplicity
            objects_high.append(tmp)
            label=1
        else:
            tmp.append(0)
            objects_low.append(tmp)
            label=0
        objects.append(tmp)
        labels.append(label)

    rows.append(row)
total_elapsed_time = round(time.clock() - start_time, 2)
print(" done in {} seconds, result = {}".format(total_elapsed_time, result))
```

```python
print(len(objects_high))
print(len(objects_low))
print(len(rows))
print(objects_high[0])
print(labels.count(1))
print(labels.count(0))
print(labels[1:1000].count(1))
```
